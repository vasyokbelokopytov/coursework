# Теоритична частина
Метод BFGS - ітераційний метод чисельної оптимізації, названий на честь його дослідників: Broyden, Fletcher, Goldfarb, Shanno. Відноситься до класу квазіньютоновскіх методів. На відміну від ньютонівських методів в квазіньютоновскіх не вираховується безпосередньо гессіан функції, тобто немає необхідності знаходити часткові похідні другого порядку. Замість цього гессіан обчислюється наближено, виходячи із зроблених до цього кроків.

В якості функції розглядається степенева функція $$f(x)=(10(x_1 - x_2)^2 + (x_1 - 1)^2)^4$$

```python
def f(x):
    [x1, x2] = x
    return (10 * ((x1 - x2) ** 2) + (x1 - 1) ** 2) ** 4

x0 = [-1.2, 0]
H = 0.00001
eps = 0.001
```

```python
def sven(x, S, alpha=0.1, printable=True):
    [x1, x2] = x
    [S1, S2] = S
    
    delta_lambda = alpha * ( (x1 ** 2 + x2 ** 2) ** (1 / 2) ) / ( (S1 ** 2 + S2 ** 2) ** (1 / 2) )
    
    lambdas = [0]
    func_values =[f([x1, x2])]
    
    direction = 1
    direction_approved = False
    k = 0
    
    def func_value(lmbd):
        return f([x1 + lmbd * S1, x2 + lmbd * S2])
        
    if printable:
        print('Метод Свенна:\n')

    while True:
        next_lambda = lambdas[k] + direction * delta_lambda * (2 ** k)
        
        if func_value(next_lambda) > func_values[k] and direction_approved == False:
            direction = -direction
            direction_approved = True
            continue
        
        direction_approved = True

        lambdas.append(next_lambda)
        func_values.append(func_value(next_lambda))
        k += 1
        
        if printable:
            print(f'Lambda{k} = {lambdas[k]}, f(Lambda{k}) = {func_values[k]}')
        
        if func_values[k] > func_values[k-1] and direction_approved == True:
            result_lambda = (lambdas[k-1] + lambdas[k]) / 2
            interval = [lambdas[k-2], result_lambda]
            if printable:
                print(f'\nИнтервал неопределенности: {interval}\n\n')
            return interval
```

```python
def golden(x, S, eps, sven_param, printable=True):
    [x1, x2] = x
    [S1, S2] = S
    
    def func_value(lmbd):
        return f([x1 + lmbd * S1, x2 + lmbd * S2])
    
    interval = sven(x, S, alpha=sven_param, printable=printable)
    [a, b] = interval
    
    lambda_1 = a + 0.382 * (b - a)
    lambda_2 = a + 0.618 * (b - a)
    
    if printable:
        print('Метод золотого сечения:\n')
        print(f'Lambda_1 = {lambda_1}, f(Lambda_1) = {func_value(lambda_1)}')
        print(f'Lambda_2 = {lambda_2}, f(Lambda_2) = {func_value(lambda_2)}\n')
    
    while abs(b - a) / 2 >= eps:
        
        if func_value(lambda_1) > func_value(lambda_2):
            a = lambda_1
            lambda_1 = lambda_2
            lambda_2 = b - (lambda_1 - a)
            
        else:
            b = lambda_2
            lambda_2 = lambda_1
            lambda_1 = a + (b - lambda_2)
            
            
        if printable:
            print(f'Lambda_1 = {lambda_1}, f(Lambda_1) = {func_value(lambda_1)}')
            print(f'Lambda_2 = {lambda_2}, f(Lambda_2) = {func_value(lambda_2)}\n')
    
    lambda_opt = (a + b) / 2
    if printable:
        print(f'Lambda_opt = {lambda_opt}')
    return lambda_opt
    
```

```python
def dsk_powell(x, S, eps, sven_param, printable=True):
    [x1, x2] = x
    [S1, S2] = S
    
    def func_value(lmbd):
        return f([x1 + lmbd * S1, x2 + lmbd * S2])
    
    interval = sven(x, S, alpha=sven_param, printable=printable)
    
    [lambda_1, lambda_3] = interval
    lambda_2 = (lambda_1 + lambda_3) / 2
    
    lambda_asterisk = lambda_2 + ((lambda_3 - lambda_2) * (func_value(lambda_1) - func_value(lambda_3)) / 
                                  (2 * (func_value(lambda_1) - 2 * func_value(lambda_2) + func_value(lambda_3))))
    
    func_lambda_1 = func_value(lambda_1)
    func_lambda_2 = func_value(lambda_2)
    func_lambda_3 = func_value(lambda_3)
    func_lambda_asterisk = func_value(lambda_asterisk)
    
    while True:
        if (abs(func_lambda_2 - func_lambda_asterisk) <= eps and abs(lambda_2 - lambda_asterisk) <= eps):
            if printable:
                print(f'Lambda_opt = {lambda_asterisk}')
            return lambda_asterisk
        
        if func_lambda_asterisk < func_lambda_2:
            if lambda_asterisk > lambda_2:
                lambda_1, lambda_2 = lambda_2, lambda_asterisk
                func_lambda_1, func_lambda_2 = func_lambda_2, func_lambda_asterisk
            else:
                lambda_2, lambda_3 = lambda_asterisk, lambda_2
                func_lambda_2, func_lambda_3 = func_lambda_asterisk, func_lambda_2
        else:
            if lambda_asterisk > lambda_2:
                lambda_3 = lambda_asterisk
                func_lambda_3 = func_lambda_asterisk
            else:
                lambda_1 = lambda_asterisk
                func_lambda_1 = func_lambda_asterisk
        
        alpha_1 = (func_lambda_2 - func_lambda_1) / (lambda_2 - lambda_1)
        alpha_2 = (((func_lambda_3 - func_lambda_1) / (lambda_3 - lambda_1)) -
                    alpha_1) / (lambda_3 - lambda_2)
        
        lambda_asterisk = ((lambda_1 + lambda_2) / 2) - (alpha_1 / (2 * alpha_2))
```

```python
def get_h(x, n):
    result = []
    for i in range(len(x)):
        if i==n:
            result.append(H)
        else:
            result.append(0)
    return result
    
def center_diff(x, n): 
    h = get_h(x, n)
    return (f([x[0] + h[0], x[1] + h[1]]) - f([x[0] - h[0], x[1] - h[1]]))/(2*h[n])

def left_diff(x, n):
    h = get_h(x, n)
    return (f(x) - f([x[0] - h[0], x[1] - h[1]]))/(h[n])

def right_diff(x, n): 
    h = get_h(x, n)
    return (f([x[0] + h[0], x[1] + h[1]]) - f(x))/(h[n])

def right_sqr_diff(x, n):
    h = get_h(x, n)
    return ( right_diff([x[0] + h[0], x[1] + h[1]], n) - right_diff([x[0], x[1]], n) ) / h[n]

def left_sqr_diff(x, n):
    h = get_h(x, n)
    return ( left_diff([x[0], x[1]], n) - left_diff([x[0] - h[0], x[1] - h[1]], n) ) / h[n]

def center_sqr_diff(x, n): 
    return (left_sqr_diff([x[0], x[1]], n) + right_sqr_diff([x[0], x[1]], n)) / 2

def center_mixed_diff(x):
    return (center_diff([x[0], x[1]+H], 0) - center_diff([x[0], x[1]], 0)) / H

def left_mixed_diff(x):
    return (left_diff([x[0], x[1]+H], 0) - left_diff([x[0], x[1]], 0)) / H

def right_mixed_diff(x):
    return (right_diff([x[0], x[1]+H], 0) - right_diff([x[0], x[1]], 0)) / H


def gradient(x):
    return [center_diff(x, i) for i in range(len(x))]
```
